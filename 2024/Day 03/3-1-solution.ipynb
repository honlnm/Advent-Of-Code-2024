```python
import re
```

```python
myRe = r"mul\(\d{1,3}\,\d{1,3}\)"
full_list = []
with open("3-puzzle_input.txt", "r") as file:
    for line in file:
        list = re.findall(myRe, line)
        full_list += list

print(full_list)
```

```python
result = 0
for item in full_list:
    numberRe = r"\d{1,3}"
    numbers = re.findall(numberRe, item)
    mul = int(numbers[0]) * int(numbers[1])
    result += mul

print(result)
```
