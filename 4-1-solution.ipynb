```python
def word_search(grid, word):
    rows, cols = len(grid), len(grid[0])
    word_length = len(word)
    count = 0

     # Define directions: (row_step, col_step)
    directions = [
        (0, 1),   # Horizontal right
        (0, -1),  # Horizontal left
        (1, 0),   # Vertical down
        (-1, 0),  # Vertical up
        (1, 1),   # Diagonal down-right
        (-1, -1), # Diagonal up-left
        (1, -1),  # Diagonal down-left
        (-1, 1)   # Diagonal up-right
    ]

    def is_valid(row, col):
        return 0 <= row < rows and 0 <= col < cols

    def search_from(start_row, start_col, direction):
        row_step, col_step = direction
        for i in range(word_length):
            row, col = start_row + i * row_step, start_col + i * col_step
            if not is_valid(row, col) or grid[row][col] != word[i]:
                return False
        return True

    # Loop through the grid and try to match the word in all directions
    for row in range(rows):
        for col in range(cols):
            if grid[row][col] == word[0]:
                for direction in directions:
                    if search_from(row, col, direction):
                        count +=1
    return count
```

```python
def read_file(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
        grid = [line.strip().split() for line in lines]
    return grid
```
