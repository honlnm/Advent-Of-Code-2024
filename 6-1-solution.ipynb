```python
def change_direction(symbol):
    if symbol == "^":
        return ">"
    if symbol == ">":
        return "V"
    if symbol == "V":
        return "<"
    if symbol == "<":
        return "^"
```

```python
def move_free_symbol(grid, row, col, symbol):
    # Validate the symbol
    if symbol not in ["^", ">", "V", "<"]:
        raise ValueError(f"Invalid symbol: {symbol}")

    # Validate the starting position
    if not (0 <= row < len(grid)) or not (0 <= col < len(grid[0])):
        raise IndexError(f"Starting position ({row}, {col}) is out of bounds")

    # Remove the symbol from the current position
    grid[row][col] = "X"

    # Calculate the new position
    if symbol == "^" and row > 0:
        row -= 1
    elif symbol == ">" and col < len(grid[0]) - 1:
        col += 1
    elif symbol == "V" and row < len(grid) - 1:
        row += 1
    elif symbol == "<" and col > 0:
        col -= 1
    else:
        # If movement is not possible, retain the symbol's position
        return grid

    # Place the symbol at the new position
    grid[row][col] = symbol

    return [grid, row, col]
```

```python
def next_symbol(grid, row, col, symbol):
    if symbol == "^" and row > 0:
        row -= 1
    elif symbol == ">" and col < len(grid[0]) - 1:
        col += 1
    elif symbol == "V" and row < len(grid) - 1:
        row += 1
    elif symbol == "<" and col > 0:
        col -= 1
    return grid[row][col]
```

```python
def gaurd_path(grid, curr_row, curr_col, symbol, count):
    ns = next_symbol(grid, curr_row, curr_col, symbol)
    if ns == "X":
        return count
    elif ns == "#":
        symbol = change_direction(grid[curr_row][curr_col])
        resp = move_free_symbol(grid, curr_row, curr_col, symbol)
        grid = resp[0]
        row = resp[1]
        col = resp[2]
        count = int(count) + 1
        gaurd_path(grid, row, col, symbol, count)
    elif ns == ".":
        resp = move_free_symbol(grid, curr_row, curr_col, symbol)
        grid = resp[0]
        row = resp[1]
        col = resp[2]
        count = int(count) + 1
        gaurd_path(grid, row, col, symbol, count)
    else:
        symbol = change_direction(grid[curr_row][curr_col])
        resp = move_free_symbol(grid, curr_row, curr_col, symbol)
        grid = resp[0]
        row = resp[1]
        col = resp[2]
        count = int(count) + 1
        gaurd_path(grid, curr_row, curr_col, symbol, count)
        

count = 0
print(gaurd_path(grid, curr_row, curr_col, symbol, count))
```
