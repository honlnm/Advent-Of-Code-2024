```python
class Move:
    def __init__(self, grid):
        self.grid = grid
        self.visited = set() # To track unique spaces visited
        self.vertical_max_ind = len(grid) - 1
        self.horizontal_max_ind = len(grid[0]) - 1

        # Directions
        self.north = [0, -1]
        self.east = [1, 0]
        self.south = [0, 1]
        self.west = [-1, 0]

    def move_north(self, gridpoint):
        new_gridpoint = [x + y for x, y in zip(gridpoint, self.north)]
        print("N: ", gridpoint, new_gridpoint)
        if new_gridpoint[1] < 0:  # Out of bounds
            self.visited.add(tuple(gridpoint))
            return self.visited
        elif self.grid[new_gridpoint[0]][new_gridpoint[1]] == "#":  # Blocked
            self.visited.add(tuple(gridpoint))
            return self.move_east(gridpoint)
        else:
            self.visited.add(tuple(gridpoint))
            return self.move_north(new_gridpoint)

    def move_east(self, gridpoint):
        new_gridpoint = [x + y for x, y in zip(gridpoint, self.east)]
        print("E: ", gridpoint, new_gridpoint)
        if new_gridpoint[0] > self.vertical_max_ind:  # Out of bounds
            self.visited.add(tuple(gridpoint))
            return self.visited
        elif self.grid[new_gridpoint[0]][new_gridpoint[1]] == "#":  # Blocked
            self.visited.add(tuple(gridpoint))
            return self.move_south(gridpoint)
        else:
            self.visited.add(tuple(gridpoint))
            return self.move_east(new_gridpoint)

    def move_south(self, gridpoint):
        new_gridpoint = [x + y for x, y in zip(gridpoint, self.south)]
        print("S: ", gridpoint, new_gridpoint)
        if new_gridpoint[1] > self.horizontal_max_ind:  # Out of bounds
            self.visited.add(tuple(gridpoint))
            return self.visited
        elif self.grid[new_gridpoint[0]][new_gridpoint[1]] == "#":  # Blocked
            self.visited.add(tuple(gridpoint))
            return self.move_west(gridpoint)
        else:
            self.visited.add(tuple(gridpoint))
            return self.move_south(new_gridpoint)

    def move_west(self, gridpoint):
        new_gridpoint = [x + y for x, y in zip(gridpoint, self.west)]
        print("W: ", gridpoint, new_gridpoint)
        if new_gridpoint[0] < 0:  # Out of bounds
            self.visited.add(tuple(gridpoint))
            return self.visited
        elif self.grid[new_gridpoint[0]][new_gridpoint[1]] == "#":  # Blocked
            self.visited.add(tuple(gridpoint))
            return self.move_north(gridpoint)
        else:
            self.visited.add(tuple(gridpoint))
            return self.move_west(new_gridpoint)
```

```python
new_instance = Move(grid)
unique_gridpoints = new_instance.move_north(gridpoint)
print(unique_gridpoints)

#135 is too low
```
